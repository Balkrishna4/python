```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# first we read the image ,to continue we used cv2
```

```python
image = cv2.imread('bottles.jpg')
plt.imshow(image)
```

```python
# in order to reduced  down noise , weight ,size of the object
#to process the image faster and smoother we uesd grey scale
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray , cmap="gray")
```

```python
#to make better use we will make the image blur 
# if we will don't do it ,it will unable to count the task
```

```python
blur = cv2.GaussianBlur(gray , (11,11),0)
plt.imshow(blur ,cmap="gray") 
```

```python
#we have to identiy edge of the blur images
#for these use candy identification of edge 
#3=kernel size for soble operation
```

```python
canny = cv2.Canny(blur, 50,240.40,3)
plt.imshow(canny , cmap="gray")
```

```python
dilated = cv2.dilate(canny, (1,1), iterations=0)
plt.imshow(dilated,cmap="gray")
```

```python
(cnt,hierarchy) = cv2.findContours(dilated.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
cv2.drawContours(rgb,cnt,-1,(0,255,0),2)
print("Bottles in the images is: " ,len(cnt))
plt.imshow(rgb)
```
